```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
num_students = 500
```

```python
study_hours = np.random.normal(4, 1.5, num_students).clip(0, 10)
social_media_hours = np.random.normal(3, 1.5, num_students).clip(0, 8)
youtube_hours = np.random.normal(2, 1, num_students).clip(0, 6)
gaming_hours = np.random.normal(1.5, 1, num_students).clip(0, 5)
sleep_hours = np.random.normal(7, 1, num_students).clip(4, 10)
phone_unlocks = np.random.randint(20, 120, num_students)
assignments_completed = np.random.randint(0, 10, num_students)
```

```python
df = pd.DataFrame({
    "study_hours": study_hours,
    "social_media_hours": social_media_hours,
    "youtube_hours": youtube_hours,
    "gaming_hours": gaming_hours,
    "sleep_hours": sleep_hours,
    "phone_unlocks": phone_unlocks,
    "assignments_completed": assignments_completed
})
```

```python
df["distraction_score"] = (
    df["social_media_hours"] * 2 +
    df["youtube_hours"] * 1.5 +
    df["gaming_hours"] * 2 +
    df["phone_unlocks"] * 0.05 -
    df["study_hours"] * 1.5
)
```
